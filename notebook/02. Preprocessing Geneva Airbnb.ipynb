```python
clean_df = df[df.first_review.notnull() & df.last_review.notnull() & df.review_scores_rating.notnull() & df.review_scores_accuracy.notnull() & df.review_scores_accuracy.notnull() & df.review_scores_cleanliness.notnull() & df.review_scores_checkin.notnull() & df.review_scores_communication.notnull() & df.review_scores_location.notnull() & df.review_scores_value.notnull()]
```

```python
clean_df.drop(['bathrooms', 'neighbourhood_group_cleansed', 'calendar_updated', 'license'], axis=1, inplace=True)
```

```python
# fill NaN with "" for string values
clean_df[['listing_url', 'name', 'description', 'neighborhood_overview', 'picture_url', 'host_url', 'host_name', 
    'host_location', 'host_about', 'host_response_time', 'host_response_rate', 'host_acceptance_rate', 
    'host_is_superhost', 'host_thumbnail_url', 'host_picture_url', 'host_neighbourhood', 'host_has_profile_pic', 
    'host_identity_verified', 'neighbourhood', 'neighbourhood_cleansed', 
    'property_type', 'room_type', 'bathrooms_text', 'price', 'first_review', 
    'last_review']] = clean_df[['listing_url', 'name', 'description', 'neighborhood_overview', 'picture_url', 'host_url', 
    'host_name', 'host_location', 'host_about', 'host_response_time', 'host_response_rate', 'host_acceptance_rate', 
    'host_is_superhost', 'host_thumbnail_url', 'host_picture_url', 'host_neighbourhood', 'host_has_profile_pic', 
    'host_identity_verified', 'neighbourhood', 'neighbourhood_cleansed',  
    'property_type', 'room_type', 'bathrooms_text', 'price', 'first_review', 
 'last_review']].fillna(value=" ")
```

```python
# fill NaN with 0.0 for float values
clean_df[['latitude', 'longitude', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 
    'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value', 
    'reviews_per_month']] = clean_df[['latitude', 'longitude', 'review_scores_rating', 'review_scores_accuracy', 
    'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 
    'review_scores_value', 'reviews_per_month']].fillna(value=0.0)
```

```python
# fill NaN with 0 for integer values
clean_df[['minimum_nights', 'maximum_nights', 'maximum_minimum_nights', 'minimum_minimum_nights', 'maximum_minimum_nights', 
    'minimum_maximum_nights', 'maximum_maximum_nights', 'minimum_nights_avg_ntm', 
    'maximum_nights_avg_ntm', 'bedrooms', 'beds']] = clean_df[['minimum_nights', 'maximum_nights', 'maximum_minimum_nights', 
    'minimum_minimum_nights', 'maximum_minimum_nights', 'minimum_maximum_nights', 'maximum_maximum_nights', 
    'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 'bedrooms', 'beds']].fillna(value=0)
```

```python
clean_df.to_csv('../data/Airbnb/Geneva/clean_listings.csv', index=False)
```
